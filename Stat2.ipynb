```python
# A Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable taking on a particular value. 
# The PDF defines the probability distribution of a continuous random variable, where the area under the curve over an interval represents 
# the probability that the random variable falls within that interval. The total area under the PDF curve is equal to 1 (since the 
# total probability is 100%). The probability of the random variable taking any specific value is technically 0, but the probability 
# over an interval can be computed as the area under the curve over that interval.
# Types of Probability Density Functions
# Uniform Distribution is a probability distribution where all outcomes are equally likely within a given range. The PDF of a uniform distribution is flat and constant within the range [a,b][a, b][a,b].
# B UC: Random Number Generation: A business might use a uniform distribution to model random events like the selection of a random employee 
# for a task, or when conducting a randomized trial where all options are equally likely.

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the uniform distribution
a = 0
b = 10

# Generate a sample of random numbers from the uniform distribution
x = np.linspace(a, b, 1000)
pdf = np.ones_like(x) / (b - a)  # PDF is constant over the range

# Plot the PDF
plt.plot(x, pdf, label="Uniform Distribution", color="blue")
plt.fill_between(x, pdf, alpha=0.3, color="blue")
plt.title("Uniform Distribution PDF")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Normal Distribution (Gaussian) - is one of the most commonly used probability distributions. It is symmetric and bell-shaped. 
# It is defined by its mean (µ) and standard deviation (σ). The probability density decreases as the value moves away from the mean.
# B UC: Employee Heights or Salaries: Many natural phenomena (e.g., heights of employees in a company, annual income distribution) follow 
# a normal distribution. A company can use the normal distribution to model such data and assess risks or set benchmarks.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate data points
x = np.linspace(-5, 5, 1000)
pdf = norm.pdf(x, mu, sigma)

# Plot the PDF
plt.plot(x, pdf, label="Normal Distribution", color="green")
plt.fill_between(x, pdf, alpha=0.3, color="green")
plt.title("Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Exponential Distribution - The Exponential Distribution models the time between events in a Poisson process, 
# which describes events happening at a constant rate. It is often used to model waiting times or lifetimes of objects.
# B UC: Customer Wait Time: In a call center, the time between customer calls or the time between purchases in an e-commerce setting 
# can follow an exponential distribution. Companies can use it to predict customer wait times or stock replenishment.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameter for the exponential distribution
lambda_param = 1  # Rate parameter

# Generate data points
x = np.linspace(0, 10, 1000)
pdf = expon.pdf(x, scale=1/lambda_param)

# Plot the PDF
plt.plot(x, pdf, label="Exponential Distribution", color="red")
plt.fill_between(x, pdf, alpha=0.3, color="red")
plt.title("Exponential Distribution PDF")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Log-Normal Distribution is a distribution of a variable whose logarithm is normally distributed. 
# It is often used to model distributions that are skewed and where values are always positive.
# B UC: Stock Prices: The prices of stocks or financial assets often follow a log-normal distribution, 
# where the daily returns are normally distributed, but the actual price is positively skewed.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parameters for the log-normal distribution
mu = 0  # Mean of the log of X
sigma = 1  # Standard deviation of the log of X

# Generate data points
x = np.linspace(0.01, 10, 1000)
pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))

# Plot the PDF
plt.plot(x, pdf, label="Log-Normal Distribution", color="purple")
plt.fill_between(x, pdf, alpha=0.3, color="purple")
plt.title("Log-Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Beta Distribution is a continuous probability distribution defined on the interval [0,1][0, 1][0,1]. 
# It is often used to model the distribution of random variables that represent proportions or probabilities.
# B UC: Market Share Estimation: The Beta distribution is often used in Bayesian statistics, such as estimating the 
# probability distribution of a company's market share or conversion rates in A/B testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters for the beta distribution
alpha = 2
beta_param = 5

# Generate data points
x = np.linspace(0, 1, 1000)
pdf = beta.pdf(x, alpha, beta_param)

# Plot the PDF
plt.plot(x, pdf, label="Beta Distribution", color="orange")
plt.fill_between(x, pdf, alpha=0.3, color="orange")
plt.title("Beta Distribution PDF")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Gamma Distribution is a family of continuous probability distributions that generalizes the exponential distribution. 
# It is used to model waiting times or lifetimes where events occur at different rates.
# B UC: Insurance Claims: In insurance, the time between claims might follow a Gamma distribution, where the shape parameter determines the number of claims in a given period.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parameters for the gamma distribution
k = 2  # Shape parameter
theta = 2  # Scale parameter

# Generate data points
x = np.linspace(0, 15, 1000)
pdf = gamma.pdf(x, k, scale=theta)

# Plot the PDF
plt.plot(x, pdf, label="Gamma Distribution", color="brown")
plt.fill_between(x, pdf, alpha=0.3, color="brown")
plt.title("Gamma Distribution PDF")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Mean Squared Error (MSE) is a common metric used to measure the performance of regression models, quantifying the difference between 
# predicted and actual values. MSE is the average of the squares of the errors (the differences between predicted and actual values). 
# It provides a measure of how well a model fits the data.
# Types of Mean Squared Error (MSE)
# Basic Mean Squared Error (MSE) MSE is the basic form of the error metric. It measures the average of the squared differences between 
# the actual and predicted values. MSE heavily penalizes larger errors due to squaring the differences.
# Characteristics: Sensitive to outliers. The smaller the MSE, the better the model.
# B UC: Predicting Housing Prices: MSE can be used to evaluate how well a regression model predicts housing prices based on features 
# such as location, square footage, etc.

from sklearn.metrics import mean_squared_error
import numpy as np

# Actual values (y) and predicted values (y_pred)
y_actual = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

# Calculate MSE
mse = mean_squared_error(y_actual, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
# Root Mean Squared Error (RMSE) RMSE is the square root of the Mean Squared Error. It provides an error metric in the 
# same units as the data, making it more interpretable than MSE.
# Characteristics: RMSE is more interpretable than MSE because it gives the error in the same unit as the original data.
# Like MSE, it is also sensitive to outliers.
# B UC: Predicting Stock Prices: RMSE can be used in stock price prediction models to assess the performance of a regression model. 
# The smaller the RMSE, the closer the model's predictions are to actual stock prices.

# Calculate RMSE
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse:.2f}")
```

```python
# Mean Absolute Error (MAE) MAE is the average of the absolute differences between the predicted and actual values.
# Unlike MSE, MAE does not square the differences, so it is less sensitive to large errors (outliers).
# Characteristics: Less sensitive to outliers compared to MSE or RMSE. Provides a more intuitive understanding of the average error 
# (the average distance between predictions and actual values).
# B UC: Delivery Times: For a logistics company, MAE could be used to evaluate how accurate delivery time predictions are 
# across a number of deliveries.

from sklearn.metrics import mean_absolute_error

# Calculate MAE
mae = mean_absolute_error(y_actual, y_pred)
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
# R-squared (Coefficient of Determination) R-squared (R²) is a statistical measure that represents the proportion of the variance 
# in the dependent variable that is predictable from the independent variables. It is often used alongside MSE, RMSE, and MAE 
# to evaluate how well the regression model fits the data.
# Characteristics: R² ranges from 0 to 1: 1 indicates a perfect fit (model explains all the variability). 
# 0 indicates that the model does not explain any of the variability.
# B UC: Sales Forecasting: R² can be used to evaluate how well a sales forecasting model explains variations in future sales 
# based on features like seasonality, promotions, etc.

from sklearn.metrics import r2_score

# Calculate R²
r_squared = r2_score(y_actual, y_pred)
print(f"R-squared: {r_squared:.2f}")
```

```python
# Z Test: The Z-test is a statistical test used to determine whether there is a significant difference between the sample mean 
# and the population mean (or between two sample means) under the assumption that the data follows a normal distribution. 
# Z-tests are based on the Z-distribution (standard normal distribution), and they are commonly used when the sample size is large 
# or the population variance is known.
# The Z-test is useful for hypothesis testing, where we typically have the following hypothesis:
# •	Null Hypothesis (H₀): The sample mean is equal to the population mean (or the two sample means are equal).
# •	Alternative Hypothesis (H₁): The sample mean is different from the population mean (or the two sample means are different).
# Types of Z-Tests
# One-Sample Z-Test - A One-Sample Z-Test is used when you want to test if the mean of a sample is significantly different from a known population mean. The population standard deviation is known or can be assumed based on prior knowledge.
# Hypothesis:•	Null Hypothesis (H₀): The sample mean is equal to the population mean. 
# Alternative Hypothesis (H₁): The sample mean is not equal to the population mean (two-tailed test) 
# or is greater than/less than the population mean (one-tailed test).
# B UC: Customer Satisfaction: Testing if the average customer satisfaction score for a sample of customers is significantly different 
# from the company's target mean satisfaction score.

import numpy as np
from scipy import stats

# Given data
sample_data = np.array([48, 50, 52, 49, 51, 53, 47, 50, 51, 49])
population_mean = 50
population_std = 2
n = len(sample_data)

# Sample mean
sample_mean = np.mean(sample_data)

# Z-test calculation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
```

```python
# Two-Sample Z-Test: A Two-Sample Z-Test is used to compare the means of two independent samples to determine if there is a significant 
# difference between them. This test assumes that both samples come from normal distributions with known population variances.
# Hypothesis: Null Hypothesis (H₀): The means of the two populations are equal. 
# Alternative Hypothesis (H₁): The means of the two populations are not equal.
# B UC: Marketing Campaign Effectiveness: Comparing the mean sales before and after a marketing campaign to check 
# if there is a significant difference between the two periods.

# Data for two samples
sample_data_1 = np.array([48, 50, 52, 49, 51, 53, 47, 50, 51, 49])
sample_data_2 = np.array([55, 57, 58, 56, 59, 60, 55, 56, 57, 58])

# Known population standard deviations
population_std_1 = 2
population_std_2 = 2

# Sample sizes
n1 = len(sample_data_1)
n2 = len(sample_data_2)

# Sample means
mean_1 = np.mean(sample_data_1)
mean_2 = np.mean(sample_data_2)

# Z-test calculation for two independent samples
z_score = (mean_1 - mean_2) / np.sqrt((population_std_1**2 / n1) + (population_std_2**2 / n2))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
```

```python
# Z-Test for Proportions - The Z-Test for Proportions is used when comparing the proportions of a categorical variable between two groups. 
# It’s used to test if the proportions of a particular outcome are significantly different between two groups.
# Hypothesis: Null Hypothesis (H₀): The proportions of the two groups are equal. 
# Alternative Hypothesis (H₁): The proportions of the two groups are not equal.
# B UC: Customer Conversion Rate: A company wants to test if the conversion rate (proportion of website visitors who make a purchase) 
# is significantly different between two marketing strategies.

# Data for two groups: successes and total sample sizes
x1 = 90  # Number of successes in group 1
n1 = 150  # Sample size of group 1

x2 = 85  # Number of successes in group 2
n2 = 140  # Sample size of group 2

# Pooled proportion
P = (x1 + x2) / (n1 + n2)

# Sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Z-test for proportions
z_score = (p1 - p2) / np.sqrt(P * (1 - P) * (1/n1 + 1/n2))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
```

```python
# T-Test - The T-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups 
# or if the mean of a single group differs from a known value. It is commonly used when the sample size is small 
# or when the population standard deviation is unknown.
# Types of T-Tests
# One-Sample T-Test A One-Sample T-Test is used to compare the mean of a single sample to a known population mean when the 
# population standard deviation is unknown. This test assumes that the data is normally distributed.
# Hypothesis: Null Hypothesis (H₀): The sample mean is equal to the population mean. 
# Alternative Hypothesis (H₁): The sample mean is not equal to the population mean (two-tailed) or is greater than/less than the population mean (one-tailed).
# B UC: Employee Performance: A company might want to test if the average score of a sample of employees on a performance evaluation is significantly different from a target average score.

import numpy as np
from scipy import stats

# Sample data
sample_data = np.array([48, 50, 52, 49, 51, 53, 47, 50, 51, 49])
population_mean = 50

# Sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# T-test calculation
t_statistic, p_value = stats.ttest_1samp(sample_data, population_mean)

print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")
```

```python
# Two-Sample T-Test (Independent T-Test) A Two-Sample T-Test is used to compare the means of two independent samples to determine 
# if there is a significant difference between them. It is appropriate when you are comparing the means of two distinct groups 
# (e.g., two different treatment groups).
# Hypothesis: Null Hypothesis (H₀): The means of the two populations are equal.
# •	Alternative Hypothesis (H₁): The means of the two populations are not equal.
# B UC: Comparing Marketing Strategies: A company may want to compare the sales performance of two different marketing strategies by 
# comparing the average sales of two independent groups.

# Sample data for two independent groups
group1 = np.array([48, 50, 52, 49, 51, 53, 47, 50, 51, 49])
group2 = np.array([55, 57, 58, 56, 59, 60, 55, 56, 57, 58])

# Two-sample t-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")
```

```python
# Paired T-Test - A Paired T-Test is used when you have two related samples or measurements 
# (e.g., measurements taken before and after an intervention on the same subjects). 
# This test helps determine if there is a statistically significant difference between the two related groups.
# Hypothesis: Null Hypothesis (H₀): The mean difference between the two related groups is zero (i.e., no significant change).
# •	Alternative Hypothesis (H₁): The mean difference between the two related groups is not zero.
# B UC: Product Testing: A company might want to test if a new product design leads to improved customer satisfaction scores by comparing scores before and after the product launch using a paired t-test.

# Paired data (before and after treatment)
before = np.array([48, 50, 52, 49, 51, 53, 47, 50, 51, 49])
after = np.array([55, 57, 58, 56, 59, 60, 55, 56, 57, 58])

# Paired t-test
t_statistic, p_value = stats.ttest_rel(before, after)

print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")
```
