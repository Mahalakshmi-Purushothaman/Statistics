```python
# A hypothesis is a testable statement or prediction about a population parameter. It is an essential part of statistical inference, 
# as it helps to draw conclusions based on sample data. In statistical testing, we perform hypothesis tests to determine whether 
# there is enough evidence in a sample to support a particular claim or hypothesis about the population.
# Example Hypothesis: H‚ÇÄ: The mean number of transactions before the campaign is equal to the mean number of transactions after the campaign.

from scipy import stats
import numpy as np

# Sample data: Transactions before and after the campaign
before_campaign = np.random.normal(30, 5, 50)  # Number of transactions before the campaign
after_campaign = np.random.normal(32, 5, 50)   # Number of transactions after the campaign

# One-sample t-test to compare means
t_stat, p_value = stats.ttest_1samp(after_campaign - before_campaign, 0)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Alternative Hypothesis (H‚ÇÅ) - The alternative hypothesis (H‚ÇÅ) is a statement that contradicts the null hypothesis. 
# It suggests that there is an effect, difference, or relationship in the population.

# Example Hypothesis (Banking Business): - H‚ÇÅ: The mean number of transactions after the campaign is greater than the mean 
# number of transactions before the campaign.

# One-tailed t-test: The alternative hypothesis is that the mean after campaign is greater
t_stat, p_value = stats.ttest_1samp(after_campaign - before_campaign, 0)
p_value = p_value / 2  # Since it‚Äôs a one-tailed test

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# 3. One-Tailed Hypothesis - A one-tailed hypothesis tests if a sample statistic is either greater than or less than a population parameter. 
# It focuses on one direction of the distribution.
#B UC: The bank might want to test if the average savings account balance has increased since the introduction of a new promotional offer. 
# This would be a one-tailed test. Example Hypothesis (Banking Business):
# H‚ÇÄ: The mean savings account balance has not increased after the promotional offer.
# H‚ÇÅ: The mean savings account balance has increased after the promotional offer.

# Sample data: Savings account balances before and after the promotion
before_offer = np.random.normal(4000, 800, 50)
after_offer = np.random.normal(4200, 850, 50)

# One-tailed t-test
t_stat, p_value = stats.ttest_1samp(after_offer - before_offer, 0)
p_value = p_value / 2  # Since it‚Äôs a one-tailed test

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# 4. Two-Tailed Hypothesis - A two-tailed hypothesis tests whether a sample statistic is either significantly greater than or less 
# than a population parameter, considering both directions.
# B UC: A bank wants to determine if the average loan amount has changed after introducing a new loan product. 
# This would be a two-tailed hypothesis test because the bank is interested in both an increase or a decrease in loan amounts.
# Example Hypothesis (Banking Business):
# H‚ÇÄ: The mean loan amount before the new product is equal to the mean loan amount after the new product.
# H‚ÇÅ: The mean loan amount before the new product is not equal to the mean loan amount after the new product.

# Sample data: Loan amounts before and after the new product
before_product = np.random.normal(15000, 3000, 50)
after_product = np.random.normal(15500, 3200, 50)

# Two-tailed t-test
t_stat, p_value = stats.ttest_ind(before_product, after_product)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# 5. Simple Hypothesis - A simple hypothesis specifies a single value for a population parameter. It is precise and unambiguous.
# B UC: Use Case (Banking Business): A bank wants to test if the average transaction fee for a new account type is exactly $2.
# Example Hypothesis: H‚ÇÄ: The mean transaction fee for the new account type is $2. H‚ÇÅ: The mean transaction fee for the new account type is not $2.

# Sample data: Transaction fees for new account type
transaction_fees = np.random.normal(2, 0.5, 50)

# One-sample t-test to test if the mean fee is $2
t_stat, p_value = stats.ttest_1samp(transaction_fees, 2)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# 6. Composite Hypothesis - A composite hypothesis involves more than one possible value for the population parameter. 
# It does not specify a fixed value but rather a range of possible values.
# B UC: A bank wants to test if the average customer loan amount is within the range of $10,000 to $15,000.
# Sample data: Customer loan amounts
loan_amounts = np.random.normal(12000, 3000, 50)

# One-sample t-test to test if the mean loan amount is within the range [10000, 15000]
t_stat, p_value = stats.ttest_1samp(loan_amounts, 12000)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Let‚Äôs perform the hypothesis test in Python using the scipy.stats library for the z-test and calculating the p-value.

import numpy as np
from scipy.stats import norm

# Given data
baseline_conversion_rate = 0.05  # p0
new_conversions = 65
total_visitors = 1000
alpha = 0.05

# Calculate sample conversion rate (p_hat)
sample_conversion_rate = new_conversions / total_visitors

# Calculate the standard error
standard_error = np.sqrt(baseline_conversion_rate * (1 - baseline_conversion_rate) / total_visitors)

# Calculate the z-score
z_score = (sample_conversion_rate - baseline_conversion_rate) / standard_error

# Calculate the p-value for a one-sided test
p_value = 1 - norm.cdf(z_score)

# Output the results
print(f"Sample Conversion Rate: {sample_conversion_rate:.4f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Determine whether to reject or not reject the null hypothesis
if p_value <= alpha:
    print("Reject the null hypothesis: There is evidence that the conversion rate has increased.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to conclude an increase in conversion rate.")
```

```python
# Parametric tests are statistical tests that make certain assumptions about the parameters of the population distribution 
# from which the sample is drawn. Specifically, they assume that the data follows a specific distribution, often a normal distribution, 
# and that the underlying population parameters (like the mean and standard deviation) are known or can be estimated.
# This test is used to infer or test hypotheses about population parameters using sample data. 
# These tests are powerful when the assumptions hold true but can give misleading results when those assumptions are violated.
# Types of Parametric Tests - One-Sample Z-Test, One-Sample T-Test, Two-Sample T-Test, Paired T-Test, Analysis of Variance (ANOVA), Linear Regression
# 1. One-Sample Z-Test - The one-sample Z-test is used to determine if the sample mean is significantly different from 
# a known population mean when the population standard deviation is known.
# B UC: In a bank, you may want to test if the average transaction amount of customers has changed after a new service is introduced. 
# The known population mean transaction is $100, and the sample mean transaction for a random sample of 50 customers is $105 
# with a known standard deviation of $20.
import numpy as np
from scipy.stats import norm

# Sample data
sample_mean = 105
population_mean = 100
population_std = 20
sample_size = 50

# Z-Test calculation
z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-test statistic: {z}")
print(f"P-value: {p_value}")

# 2. One-Sample T-Test - The one-sample T-test is used when the population standard deviation is unknown, and the sample size is small 
# (typically ùëõ < 30). It compares the sample mean to a known population mean.
# B UC: A bank wants to test if the average savings account balance in a branch is $5000. A sample of 25 account balances is taken, 
# and the sample mean balance is $5200 with a sample standard deviation of $1500.

# Sample data
sample_mean = 5200
population_mean = 5000
sample_std = 1500
sample_size = 25

# T-Test calculation
t_stat, p_value = stats.ttest_1samp(np.random.normal(sample_mean, sample_std, sample_size), population_mean)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# 3. Two-Sample T-Test - The two-sample T-test is used to determine if two independent samples have significantly different means. 
# The populations are assumed to be normally distributed, and the test compares the means of two groups.# 
# B UC: A bank wants to compare the average loan amounts given to two groups of customers (one group with good credit 
# and another with poor credit). They collect a sample of 30 loan amounts for each group.

# Sample data for two groups
group_1 = np.random.normal(12000, 2000, 30)  # Good credit group
group_2 = np.random.normal(10000, 2500, 30)  # Poor credit group

# Two-Sample T-Test calculation
t_stat, p_value = stats.ttest_ind(group_1, group_2)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# 4. Paired T-Test - The paired T-test compares the means of two related groups, typically before and after treatment or intervention. 
# The data must be paired, meaning each data point in one group has a corresponding data point in the other group.
# Use Case : A bank wants to evaluate if a new customer support tool improves the customer satisfaction score. 
# It collects the scores of 20 customers before and after the tool was introduced.

# Paired sample data
before = np.random.normal(75, 10, 20)  # Customer satisfaction before
after = np.random.normal(80, 10, 20)   # Customer satisfaction after

# Paired T-Test calculation
t_stat, p_value = stats.ttest_rel(before, after)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# 5. Analysis of Variance (ANOVA) - ANOVA is used to determine if there are significant differences between the means of three or more groups. 
# It tests whether any of the group means are different, not necessarily which ones.
# B UC: A bank wants to test if the average monthly spending on credit cards differs among three customer age groups: 18-25, 26-40, and 41-60.

# Sample data for three age groups
group_1 = np.random.normal(1500, 300, 30)  # Age 18-25
group_2 = np.random.normal(1600, 350, 30)  # Age 26-40
group_3 = np.random.normal(1700, 400, 30)  # Age 41-60

# ANOVA calculation
f_stat, p_value = stats.f_oneway(group_1, group_2, group_3)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# 6. Linear Regression - models the relationship between a dependent variable and one or more independent variables. It assumes a linear relationship.
# B UC: A bank wants to predict customer loan amounts based on their credit score.

from sklearn.linear_model import LinearRegression

# Sample data: Credit score and loan amount
credit_score = np.random.normal(650, 50, 100).reshape(-1, 1)
loan_amount = credit_score * 50 + np.random.normal(1000, 100, 100)

# Linear regression
model = LinearRegression()
model.fit(credit_score, loan_amount)

print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")

# Non-parametric tests are statistical tests that do not assume a specific distribution for the population from which the sample is drawn. 
# Unlike parametric tests, they are more flexible and can be used when the assumptions of normality or known parameters 
# (like population mean or standard deviation) are not met. Non-parametric tests are also called "distribution-free tests" 
# because they don‚Äôt require assumptions about the distribution of the underlying population. These tests are useful for analyzing ordinal
# data or when data do not follow a normal distribution. Types of Non-Parametric Tests - Chi-Square Test, Mann-Whitney U Test, 
# Wilcoxon Signed-Rank Test, Kruskal-Wallis H Test, Friedman Test, Spearman Rank Correlation
# 1. Chi-Square Test -  is used to assess the association between categorical variables. It checks if the distribution of observed frequencies
# in a contingency table matches the expected distribution under the null hypothesis.
# B UC: A bank wants to see if the choice of account type (e.g., savings, checking, credit) is independent of customer age groups 
# (e.g., 18-25, 26-40, 41-60).

import scipy.stats as stats
import numpy as np

# Observed data (rows = age groups, columns = account types)
observed = np.array([[30, 40, 30],  # Age group 18-25
                     [50, 70, 50],  # Age group 26-40
                     [40, 60, 40]]) # Age group 41-60

# Chi-Square Test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)

print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")

# 2. Mann-Whitney U Test - The Mann-Whitney U Test is a non-parametric test that compares the distributions of two independent samples
# to determine if they come from the same distribution. It is the non-parametric equivalent of the independent two-sample t-test.
# B UC: A bank wants to compare the monthly spending habits of two groups of customers: those with good credit and those with poor credit. 
# It uses a sample of 30 customers from each group.

# Sample data
group_1 = np.random.normal(1500, 300, 30)  # Good credit group
group_2 = np.random.normal(1200, 350, 30)  # Poor credit group

# Mann-Whitney U Test
u_stat, p_value = stats.mannwhitneyu(group_1, group_2)

print(f"U-statistic: {u_stat}")
print(f"P-value: {p_value}")

# 3. Wilcoxon Signed-Rank Test - The Wilcoxon Signed-Rank Test is used to compare two related samples, matched pairs, 
# or repeated measurements on a single sample to determine if their population mean ranks differ. 
# It is the non-parametric counterpart to the paired t-test.
# B UC: A bank wants to compare customer satisfaction scores before and after a new mobile banking app was launched. 
# A sample of 20 customers provides satisfaction scores for both time periods.

# Paired sample data
before = np.random.normal(75, 10, 20)  # Satisfaction before
after = np.random.normal(80, 10, 20)   # Satisfaction after

# Wilcoxon Signed-Rank Test
stat, p_value = stats.wilcoxon(before, after)

print(f"Test statistic: {stat}")
print(f"P-value: {p_value}")

# 4. Kruskal-Wallis H Test - The Kruskal-Wallis H Test is used to determine if there are statistically significant differences between the 
# distributions of three or more independent groups. It is the non-parametric counterpart to one-way ANOVA.
# B UC: A bank wants to test if the average loan amounts given to customers differ across three regions: North, South, and West.
# A sample of 30 customers from each region is taken.

# Sample data for three regions
north = np.random.normal(12000, 2500, 30)
south = np.random.normal(15000, 3000, 30)
west = np.random.normal(11000, 2000, 30)

# Kruskal-Wallis H Test
h_stat, p_value = stats.kruskal(north, south, west)

print(f"H-statistic: {h_stat}")
print(f"P-value: {p_value}")

# 5. Friedman Test - The Friedman Test is used to detect differences in treatments across multiple test attempts.
# It is the non-parametric counterpart of the repeated measures ANOVA.
 #B UC: A bank wants to compare customer satisfaction scores from three different customer service channels: phone, email, and in-person. 
# Satisfaction scores are collected from 30 customers for all three channels.

# Sample data for three channels
phone = np.random.normal(70, 10, 30)
email = np.random.normal(75, 10, 30)
in_person = np.random.normal(80, 10, 30)

# Friedman Test
stat, p_value = stats.friedmanchisquare(phone, email, in_person)

print(f"Chi-square statistic: {stat}")
print(f"P-value: {p_value}")

# 6. Spearman Rank Correlation - The Spearman Rank Correlation measures the strength and direction of association between two ranked variables.
# It is a non-parametric version of the Pearson correlation, used when data are ordinal or not normally distributed.
# B UC: A bank wants to see if there is a correlation between customer satisfaction scores and their likelihood of recommending the bank to others.

# Sample data for satisfaction and likelihood to recommend
satisfaction = np.random.normal(70, 10, 30)
recommendation = np.random.normal(75, 10, 30)

# Spearman Rank Correlation
rho, p_value = stats.spearmanr(satisfaction, recommendation)

print(f"Spearman's Rank Correlation Coefficient: {rho}")
print(f"P-value: {p_value}")
```

```python
# Skewness is a measure of the asymmetry or departure from symmetry in the distribution of data. It tells you about the direction 
# and degree of the tail of the distribution curve. A distribution is said to be skewed if one of its tails is longer or fatter than the other.
# Types of Skewness - Positive Skewness (Right Skewness), Negative Skewness (Left Skewness), Zero Skewness (Symmetric Distribution), 
# 1. Positive Skewness (Right Skewness)
# A distribution is said to have positive skewness or right skewness when the right tail (larger values) is longer or more spread out than 
# the left tail. In other words, most of the data points lie on the left of the mean, and the distribution has a long right tail.
# B UC: In banking, the income distribution of customers is often positively skewed. While most customers might have lower or middle-income levels, 
# a few high-income individuals can skew the distribution to the right.
# Example Hypothesis (Banking Business): A bank may find that the loan amount distribution among customers is positively skewed, 
# indicating that while most customers take out smaller loans, a few customers take out very large loans.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

# Generating positively skewed data (most data points are on the lower end)
data = np.random.lognormal(mean=0, sigma=1, size=1000)

# Calculate skewness
skewness_value = skew(data)

# Plotting the data
sns.histplot(data, kde=True)
plt.title(f"Positive Skewness (Right Skewness) | Skewness: {skewness_value:.2f}")
plt.show()

print(f"Skewness Value: {skewness_value}")

# 2. Negative Skewness (Left Skewness) - A distribution is said to have negative skewness or left skewness when the left tail (smaller values) 
# is longer or more spread out than the right tail. This means that most of the data points lie on the right of the mean, and the distribution 
# has a long left tail.
# B UC: In banking, the loan default rates might exhibit negative skewness. Most customers may have low or no defaults, 
# but a few might have very high defaults, causing the distribution to skew left.
# Example Hypothesis (Banking Business): A bank might analyze the age distribution of customers and find that a large number of young customers 
# are concentrated on the right side, while a few elderly customers skew the age distribution to the left.

# Generating negatively skewed data (few customers with very low values)
data_negative_skew = np.random.beta(a=2, b=5, size=1000)

# Calculate skewness
skewness_value_negative = skew(data_negative_skew)

# Plotting the data
sns.histplot(data_negative_skew, kde=True)
plt.title(f"Negative Skewness (Left Skewness) | Skewness: {skewness_value_negative:.2f}")
plt.show()

print(f"Skewness Value: {skewness_value_negative}")

# 3. Zero Skewness (Symmetric Distribution) - A distribution has zero skewness if it is symmetric. This means the left and right tails 
# are of equal length. The data is evenly distributed around the mean.
# B UC:  In banking, the interest rates on loans across various products (e.g., personal loans, home loans, auto loans) might exhibit zero skewness, 
# indicating that the rates are symmetrically distributed around the mean.
# Example Hypothesis (Banking Business): A bank might analyze the average savings balance of customers and find a symmetric distribution, 
# meaning most customers have balances close to the mean. 

# Generating a symmetric normal distribution (zero skewness)
data_symmetric = np.random.normal(loc=5000, scale=1000, size=1000)

# Calculate skewness
skewness_value_symmetric = skew(data_symmetric)

# Plotting the data
sns.histplot(data_symmetric, kde=True)
plt.title(f"Zero Skewness (Symmetric Distribution) | Skewness: {skewness_value_symmetric:.2f}")
plt.show()

print(f"Skewness Value: {skewness_value_symmetric}")
```

```python
# Kurtosis is a statistical measure used to describe the distribution of data points in a dataset, specifically the "tailedness" of the distribution. 
# It indicates how heavy or light the tails of a distribution are compared to a normal distribution. 
# Kurtosis helps to understand whether extreme values (outliers) are more or less frequent in the data. 
# Kurtosis can be divided into different types based on the degree of tail heaviness.
# Types of Kurtosis: Leptokurtic (Positive Kurtosis), Platykurtic (Negative Kurtosis), Mesokurtic (Zero Kurtosis)
# 1. Leptokurtic (Positive Kurtosis) - A distribution is leptokurtic if it has heavy tails and a sharp peak at the center. 
# In simple terms, a leptokurtic distribution has more extreme outliers or data points that are far away from the mean. 
# It exhibits a high likelihood of extreme values (outliers).
# B UC: In banking, a loan default rate distribution might be leptokurtic. This means that while most customers make regular payments, 
# there are a few customers with large defaults, causing the distribution to have heavy tails.
# Example Hypothesis (Banking Business): A bank may analyze the transaction volume of customers and find that a few high-value transactions 
# (e.g., huge transfers or investments) create a distribution with heavy tails, indicating that the data is leptokurtic.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Generating leptokurtic (heavy-tailed) data
data_leptokurtic = np.random.laplace(loc=0, scale=1, size=1000)

# Calculate kurtosis
kurt_value_leptokurtic = kurtosis(data_leptokurtic)

# Plotting the data
sns.histplot(data_leptokurtic, kde=True)
plt.title(f"Leptokurtic Distribution | Kurtosis: {kurt_value_leptokurtic:.2f}")
plt.show()

print(f"Kurtosis Value: {kurt_value_leptokurtic}")

# 2. Platykurtic (Negative Kurtosis) - A distribution is platykurtic if it has light tails and a flatter peak compared to a normal distribution. 
# In a platykurtic distribution, extreme values (outliers) are less likely to occur.
# B UC: In banking, the deposit amounts might be platykurtic. Most customers may deposit small amounts, 
# and the chances of very large deposits are low. This would result in a distribution with lighter tails.
# Example Hypothesis: A bank may find that the credit card balances across customers are relatively uniform with few extreme values. 
# This would indicate platykurtic distribution because most customers have average balances and very few have exceptionally high or low balances.

# Generating platykurtic (light-tailed) data
data_platykurtic = np.random.beta(a=0.5, b=0.5, size=1000)

# Calculate kurtosis
kurt_value_platykurtic = kurtosis(data_platykurtic)

# Plotting the data
sns.histplot(data_platykurtic, kde=True)
plt.title(f"Platykurtic Distribution | Kurtosis: {kurt_value_platykurtic:.2f}")
plt.show()

print(f"Kurtosis Value: {kurt_value_platykurtic}")

# 3. Mesokurtic (Zero Kurtosis) - A distribution is mesokurtic if its kurtosis value is equal to 3, indicating that the distribution 
# has a similar peakedness and tail behavior as a normal distribution. This type of distribution is neither heavy nor light-tailed.
# B UC: In banking, average loan amounts or average transaction amounts might follow a mesokurtic distribution. 
# These distributions follow the pattern of a normal distribution, meaning the majority of data points are near the mean, 
# with a balanced distribution of outliers.
# Example Hypothesis (Banking Business): A bank may find that the interest rates on most of its loan products are normally distributed, 
# with the majority of the loans having interest rates clustered around an average value, exhibiting mesokurtic distribution.

# Generating mesokurtic (normal) data
data_mesokurtic = np.random.normal(loc=5000, scale=1000, size=1000)

# Calculate kurtosis
kurt_value_mesokurtic = kurtosis(data_mesokurtic)

# Plotting the data
sns.histplot(data_mesokurtic, kde=True)
plt.title(f"Mesokurtic Distribution | Kurtosis: {kurt_value_mesokurtic:.2f}")
plt.show()

print(f"Kurtosis Value: {kurt_value_mesokurtic}")
```
